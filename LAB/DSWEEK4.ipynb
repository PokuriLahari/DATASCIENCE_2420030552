Data Collection

```python
import pandas as pd                 # Load dataset (Titanic dataset from seaborn or CSV)
import seaborn as sns
df = sns.load_dataset("titanic")
print("Data shape:", df.shape)
print(df.head())
```

 Data Cleaning & Preprocessing

```python
# Handle missing values
df['age'].fillna(df['age'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
# Drop duplicates
df.drop_duplicates(inplace=True)
# Encode categorical variables
df = pd.get_dummies(df, columns=['sex','class','embarked'], drop_first=True)
# Feature engineering: family size
df['family_size'] = df['sibsp'] + df['parch']
print(df.head())
```

Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Before calculating correlation, ensure all relevant columns are numeric.
# The 'df' at this point might still contain object-type columns that were not
# handled by the `pd.get_dummies` in the previous data cleaning step.
# These include 'who', 'deck', 'embark_town', and 'alive'.

# Create a temporary DataFrame to perform additional encoding for correlation calculation.
df_for_corr = df.copy()

# Identify object columns that still need encoding for correlation
object_cols_to_encode = df_for_corr.select_dtypes(include='object').columns

# Apply one-hot encoding to these remaining object columns
df_for_corr = pd.get_dummies(df_for_corr, columns=object_cols_to_encode, drop_first=True)

# Also, ensure boolean columns are treated as integers for correlation.
# While pandas usually handles them, explicit conversion can be safer.
for col in df_for_corr.select_dtypes(include='bool').columns:
    df_for_corr[col] = df_for_corr[col].astype(int)

# Histogram of age
sns.histplot(df_for_corr['age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()

# Correlation matrix
corr = df_for_corr.corr() # Now, all columns should be numeric or converted booleans
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

 Data Modeling
